# Data collection


```python
import os
import requests 
import matplotlib
from yelpapi import YelpAPI 

API_KEY = os.environ.get('YELP_KEY')

ENDPOINT = 'https://api.yelp.com/v3/businesses/search'

# create a dictionary for the header to send the API key along the request to authorize ourselves
HEADERS = {'Authorization': 'Bearer %s' % API_KEY}   #take my api key to the %s string holder

#define
PARAMETER = {'term':'food',
            'location':'Riverside, CA',
            'limit':'50'}
#request
res = requests.get(url = ENDPOINT,params = PARAMETER, headers = HEADERS)
data = res.json()

import json
import pandas as pd
from pandas.io.json import json_normalize


df = json_normalize(data['businesses'])
df
```

**Note**

We can see that the reponse body from Yelp has 24 columns. However, we only need some of them such as name, star and review counts. So we will clean the data frame to fit our needs.

```python
restaurantsDF = df[['name','review_count','rating']]
restaurantsDF.plot(kind='scatter',x='review_count',y='rating')
```

**Observation**

As we can see in the scatter plot, the averge rating of restaurants around Riverside is pretty good (above 4.0) even for that one restaurant with over 2000 reviews. From this, we can conclude that the food in Riverside is great in general. However, if we inpsect the graph carefully, we also see that there are some outliers that are lower than 3.5 and at 5.0. We should inspects these as well.

```python
low = restaurantsDF[restaurantsDF['rating'] <= 3.5]
low
```

There are 2 restaurants with less than 40 review. We can drop these as it is not fair to compare them with other restaurants with more than 450 review counts.

```python
low.drop(low.index[[6,5]])
```

```python
high = restaurantsDF[restaurantsDF['rating'] >= 4.5]
high
```

Again, we should only keep the restaurants with reivew counts higher than 400

```python
high = high[high['review_count'] >= 400]
high
```

**Conclusion**

Looking at the high and low data frame, we can see that restaurants in Riverside are not famous for their American BBQ, fried chickens and Thai food but they are great in exotic/foreign grill meat (Mid East and Greek), burgers and Japanese cuisines especially sushi (Poke)
